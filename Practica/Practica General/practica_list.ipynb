## Listas

```python
numeros = [5, 10, 15, 20]

print("Imprimir por numero: ")
for n in numeros: 
    print(n)

print("Con indice: ")
for i, n in enumerate(numeros, 1):
    print(f"{i}. {n}")
```

```python
numeros = [5, 10, 15, 20]
# Imprimí el primer y el último elemento

print(numeros[0])
print(numeros[-1])
```

```python
numeros = [5, 10, 15, 20]
# Imprimí la cantidad de elementos

print(f"La lista tiene {len(numeros)} elementos")
```

```python
numeros = [3, 8, 12, 5, 9]

# Imprimí los números mayores a 6
print(list(num for num in numeros if num > 6))

# Imprimí los números pares mayores a 6
numeros_2 = [3, 8, 12, 5, 9]
print(list(num for num in numeros_2 if num % 2 == 0 and num >= 6))
```
