## Practica de recorridos y filtrados de diccionarios

```python
persona = {
    "nombre": "Ana",
    "edad": 22,
    "ciudad": "La Plata",
    "profesion": "estudiante"
}

# Imprimir las claves
print("-" * 20)
print("Claves:")
for key in persona.keys():
    print(key)

# Imprimir los valores
print("-" * 20)
print("Valores:")
for value in persona.values():
    print(value)
    
print("-" * 20)
print("Claves y valores:")
for key, value in persona.items():
    print(f"{key}: {value}")

```

```python
# Imprimir solo persona mayores de edad
edades = {
    "Ana": 22,
    "Juan": 17,
    "Pedro": 30,
    "Lucia": 15,
    "Maria": 25
}

for name, age in edades.items():
    if age >= 18:
        print(f"{name} es mayor de edad con {age} años.")
```

```python
# Calcular total de ventas
sales = {
    "lunes": 120,
    "martes": 150,
    "miercoles": 90,
    "jueves": 200,
    "viernes": 300
}
top_sales_day = ""
top_sales = -1

total = 0
for key, value in sales.items():
    if value > top_sales:
        top_sales_day = key
        top_sales = value
    total += value

print(f"El dia con mayores ventas fue {top_sales_day} con {top_sales}")
print(f"El total vendido fue {total}")
```

### Diccionarios de listas

```python
materias = {
    "matematica": [7, 8, 9],
    "programacion": [10, 9, 8],
    "ingles": [6, 7, 8]
}

# Imprimir las materias
for s in materias.keys():
    print(s)

print("-" * 20) 
print("Otra opcion con indices usando enumerate(): ")
for i, s in enumerate(materias.keys(), 1):
    print(f"{i}. {s} ")

print("-" * 20) 
print("Notas: ")

for keys, list_notas in materias.items():
    print(f"{keys}: ")
    for k in list_notas:
        print(k)
```

```python
equipos = {
    "River": ["Enzo", "Borja", "Colidio"],
    "Boca": ["Cavani", "Romero"],
    "Racing": ["Martinez", "Rojas", "Sigali", "Moreno"]
}

# Contar elementos
for key, value in equipos.items():
    print(f"{key} tiene {len(value)} jugadores")

```

```python
import random

words_dic = {
    "Fulbo" : ["independiente", "river", "boca", "racing", "huracan"],
    "Series" : ["breaking", "thrones", "jujutsu", "shingeki", "mandalorian"],
    "Top5 mejores guitarristas" : ["gardiner", "wong", "henson", "danel", "zytecki", "rafa", "poniaseisno"],
    "Placebos" : ["Nada", "Vacio", "Cero", "Nulo"],
    "Standard" : ["python", "programa", "variable", "funcion", "bucle", "cadena", "entero", "lista"]
}

# Mostrar categorias (Menu)
for i, cat in enumerate(words_dic.keys(), 1):
    print(f"{i}. {cat}")


# Mostrar palabras de una sola categorias:
cat = int(input("Ingrese la categoria: "))

# queremos obtener una lista de los elementos segun la categoria elegida.

# Obtenemos primero una lista de las categorias (Ahora podemos acceder por indice)
categories = list(words_dic.keys())
print(categories)

# Con el indice ingresado accedemos a la categoria elegida
cat_elegida = categories[cat - 1]
print(cat_elegida)

# Con el indice (Que ahora es un str) podemos acceder al diccionario
word = random.choice(words_dic[cat_elegida])
print(f"La palabra elegida es: {word}")

```
