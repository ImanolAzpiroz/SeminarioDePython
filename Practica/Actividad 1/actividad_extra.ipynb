```python
def agregar_equipo(table):

    

    while True:

        team_name = input("Ingrese el nombre del equipo: ")
        exist = False

        for team in table:
            if team["name"] == team_name:
                # El equipo existe, mostrar mensaje y salir del for de busqueda
                print(f"El equipo '{team_name}' ya existe en la tabla.")
                exist = True
                break

        # El equipo no existe, agregarlo a la tabla y salir del bucle general
        if not exist:
            new_team = {
                "name": team_name,
                "points": 0,
                "goals_for": 0,
                "goals_against": 0
            }
            table.append(new_team)
            print(f"Equipo '{team_name}' agregado a la tabla.")
            break
            
    print("Equipos actuales en la tabla:")
    print("-" * 20)
    for team in table:
        print(team["name"])
        

def mostrar_tabla(table):
    # Testeo funciones lambda:
    # lambda parametro: expresion con el parametro
    # ejemplo lambda x: x % 2 == 0   (Si no me estoy confundiendo esto deberia devolver si x es par o no)

    sorted_table = sorted(table, key=lambda x: (x['points']), reverse=True)
    print("\nTabla ordenada por puntos:")
    for team in sorted_table:
        print(f"Equipo: {team['name']}, Puntos: {team['points']}")

    # Lo que entendi de sorted()
    # Ordena una lista de elementos, un iterable segun la key
    # La key es una funcion que se aplica a cada elemento de la lista para obtener un valor de comparacion
    # En este caso la funciona lambda x: x['points'] es el valor de comparacion 


# ---- PROGRAMA PRINCIPAL ----

menu = ["Agregar equipo", "Registrar resultado", "Eliminar equipo", "Mostrar tabla", "Salir"]

# Tabla de pruebas:
table = [
    {"name": "Manchester United", "points": 20, "goals_for": 30, "goals_against": 10},
    {"name": "Liverpool", "points": 25, "goals_for": 35, "goals_against": 15},
    {"name": "Chelsea", "points": 15, "goals_for": 25, "goals_against": 20},
    {"name": "Arsenal", "points": 18, "goals_for": 28, "goals_against": 12},
    {"name": "Manchester City", "points": 30, "goals_for": 40, "goals_against": 10},
    {"name": "Tottenham", "points": 22, "goals_for": 32, "goals_against": 18},
    {"name": "Everton", "points": 12, "goals_for": 20, "goals_against": 25},
    {"name": "West Ham", "points": 10, "goals_for": 18, "goals_against": 30}
] 


while True:
    print("Seleccione una opcion: ")
    for i, option in enumerate(menu, 1):
        print(f"{i}. {option}")

    option = int(input("Ingrese su seleccion: "))

    # Agregar equipo
    if option == 1:
        agregar_equipo(table)

    if option == 2:
        print("Registrar resultado - Funcionalidad en desarrollo.")

    if option == 3:
        print("Eliminar equipo - Funcionalidad en desarrollo.")
    
    # Mostrar tabla
    if option == 4:
        mostrar_tabla(table)
    
    # Salir
    if option == 5:
        print("Saliendo del programa.")
        break






```
